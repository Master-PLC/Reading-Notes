#### 公式(1.3)的理解

公式(1.3)为：

$$ w(t+1)=w(t)+y(t)x(t) $$

**参考**：

![formula1.3](img/formula1.3.png)

分两种情况：

+ 当 $y = +1$ 时，$sign(w^{T}x)$分类错误，$w^{T}x$ 小于 $0$，说明 $w$ 和 $x$ 夹角过大，需要让 $w$ 往 $x$ 靠近，令 $w=w+(+1)*x$ 即可
+ 当 $y = -1$ 时，$sign(w^{T}x)$分类错误，$w^{T}x$ 大于 $0$，说明 $w$ 和 $x$ 夹角过小，需要让 $w$ 远离 $x$，令 $w=w+(-1)*x$ 即可

We are given a data set $\mathcal D$ of 25 training examples from an unknown target function $f:\mathcal X \to \mathcal Y$, where $\mathcal X=\mathbb R$ and $\mathcal Y= \{-1, +1\}$. (c) If $p = 0.9$, what is the probability that $S$ will produce a better hypothesis than $C$?

(d) Is there any value of $p$ for which it is more likely than not that $C$ will produce a better hypothesis than $S$?

**参考**：

(a) 利用 $S$ 算法会选择在训练集上效果较好的假设，但是并不能保证在训练集以外的部分表现的很好

(b) 这题及之后的习题假设在训练集上的 $y_n = +1$，利用 $C$ 算法得到的假设在训练集上的准确率为0，利用 $S$ 算法得到的假设在训练集上的准确率为1，但是在训练集以外的部分未知，所以C算法得到的假设仍然有可能比S算法得到的假设表现好

$$
\mathbb P(\mathbb P[f_s=f]>\mathbb P[f_c=f])
$$

因为 $f_s(x)$ 恒等于 $1$，$\mathbb P[f(x) = +1] = 0.9$，所以

$$
\mathbb P[f_s=f] = \mathbb P[f(x) = +1] = p = 0.9 \\
\mathbb P[f_c=f] = 1-\mathbb P[f_s=f] = 0.1
$$

因此

$$
\mathbb P(\mathbb P[f_s=f]>\mathbb P[f_c=f]) = \mathbb P(0.9>0.1) = 1
$$

(d) 这题是上题的一般化结果,题目实际上问的是是否存在$p$，使得

$$
\mathbb P(\mathbb P[f_s=f]>\mathbb P[f_c=f]) < 0.5
$$

因为

$$
\mathbb P[f(x) = +1] = p
$$

所以

$$
\mathbb P[f_s=f] = p，\mathbb P[f_c=f] = 1-p \\
\mathbb P(\mathbb P[f_s=f]>\mathbb P[f_c=f]) = \mathbb P(p>1-p)
$$

所以当$p<0.5$时，

$$
\mathbb P(\mathbb P[f_s=f]>\mathbb P[f_c=f]) = 0 < 0.5
$$

因此 $p<0.5$
